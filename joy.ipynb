## Interactive Widgets

Jupyter supports interactive widgets that allow adding a slider to change values and see the updated image in real-time.

Let's start with a simple example of star and see if we can make it interactive.

```python
def star(n):
    angle = 180/n
    return line() | repeat(n, rotate(angle))
```

```python
star(5)
```

Let's make it interactive using the `interactive` function. Notice that we have specified the value of `n` as an `IntSlider`.

```python
interact(star, n=IntSlider(value=5, min=1, max=30))
```

Interactive widgets makes it easy to discover the correct value of scale or angle when making complex shapes.

```python
dx = 300/4
a=s1|scale(1/4)|translate(x=0)
b=s2|scale(1/4)|translate(x=dx)
c=s3|scale(1/4)|translate(x=2*dx)
d=s4|scale(1/4)|translate(x=3*dx)

shape = a+b+c+d

shape|translate(x=-150+dx/2)
```

```python
def row(shapes):
    n=len(shapes)
    dx=300/n
    offset=-150 + dx/2
    new_shapes=[shapes[i]|scale(1/n)|translate(x=dx*i) for i in range(n)]
    return combine(new_shapes)|translate(x=offset)
```

```python
row([s1, s2])
```

```python
def super_grid(shape_maker, n):
    size = 300/n
    r = size/2
    x = -150+r
    y = -150+r
    shapes = []
    for j in range(n):
        for i in range(n):
            c = shape_maker(size)|translate(x=x, y=y)
            shapes.append(c)
            x = x+size
        y = y+size
        x = -150+r
    return combine(shapes)
```

## Super_Grid v2

```python
def super_grid_v2(shape_maker, n):
    size = 300/n
    r = size/2
    x = -150+r
    y = -150+r
    shapes = []
    for j in range(n):
        for i in range(n):
            c = shape_maker(size, row=j, column=i, n=n)|translate(x=x, y=y)
            shapes.append(c)
            x = x+size
        y = y+size
        x = -150+r
    return combine(shapes)
```

```python
def factorial(n):
    if n>1:
        return n*factorial(n-1)
    else:
        return 1
```

```python
from joy import *

def mega_circle_bowl(r, n):
    c1=circle(r=r)
    if n>1:
        c2=mega_circle_bowl(r=r/3, n=n-1)
        c3=mega_circle_bowl(r=r/3, n=n-1)|translate(y=r/1.5)|repeat(6, rotate(60))
        return c1+c2+c3
    else:
        return c1
```
